```
from keras.utils import to_categorical
```

```
ytest=to_categorical(ytest)
ytrain=to_categorical(ytrain)
```

```
def convert(dataset):
  maxlen=10
  vol=np.zeros((dataset.shape[0],maxlen,50))     # as each word is in 50 dims 
  for i  in range(len(dataset)):
    for j in range(len(line)):
      try:
        vol[i][j]=wordembed[dataset[i][j].lower()] # length upto 10 is considered
      except:
        vol[i][j]=np.zeros((50,)) # remaining  word are setted as zero
  return vol
```

```
xtrainemb=convert(xtrain)
```

```
xtestemb=convert(xtest)
```

```
print(xtrainemb.shape,xtestemb.shape)
```

```
xtrainemb[0]
```

```
from tensorflow.python.keras.layers import Flatten,Dense,LSTM,Dropout,Activation
from tensorflow.python.keras.activations import softmax,relu,sigmoid
from tensorflow.python.keras.models import Sequential
```

```
model=Sequential()

model.add(LSTM(64,input_shape=(10,50),return_sequences=True,recurrent_dropout=0.1))
model.add(Dropout(0.2))

model.add(LSTM(64,input_shape=(10,50),return_sequences=False,recurrent_dropout=0.1))
model.add(Dropout(0.5))

model.add(Dense(20))
model.add(Activation('sigmoid'))

model.add(Dense(5))
model.add(Activation('softmax'))
model.summary()
```

```
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```
model.fit(xtrainemb,ytrain,batch_size=20,epochs=50,validation_split=0.1)
```

```
model.save("myemoji.h5")
from tensorflow.python.keras.callbacks import EarlyStopping,ModelCheckpoint
```
